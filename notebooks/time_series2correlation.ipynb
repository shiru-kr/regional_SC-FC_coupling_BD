concat the time series from cortical and subcortical atlas

```python
import pandas as pd
import os

# List of CSV files in each directory
list_sch = glob.glob('/home/shir/Documents/connectomics/sch200_tian1 (copy)/HC/time_series/*-sch.csv')
list_msaI = glob.glob('/home/shir/Documents/connectomics/sch200_tian1 (copy)/HC/time_series/*-tian.csv')

# Sort files to ensure proper ordering
list_sch.sort()
list_msaI.sort()

# Iterate through the files and concatenate them
for file1, file2 in zip(list_sch, list_msaI):
    
    # Read CSV files
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)
    
    # Concatenate dataframes
    concatenated_df = pd.concat([df1, df2], axis=0)
    
    # Write concatenated dataframe to a new CSV file
    concatenated_df.to_csv(f'/home/shir/Documents/connectomics/sch200_tian1 (copy)/HC/time_series/{file1[-15:-8]}_whole_brain.csv', index=False) 
    
    print(f'Concatenated file created for {file1} and {file2}')
```

### Compute corr

Correlation

```python
list_file = glob.glob('/home/shir/Documents/connectomics/sch200_tian1 (copy)/HC/time_series/*_whole_brain.csv')
my_data = pd.DataFrame()
for i in list_file:
    df = pd.read_csv(i, index_col=0)
    time_series=df.values
    time_series=np.transpose(time_series)
    labels = df.index.values.tolist()
    correlation_measure = ConnectivityMeasure(
        kind="correlation",
        standardize=True,
    )
    correlation_matrix = correlation_measure.fit_transform([time_series])[0]
    corre =  pd.DataFrame(correlation_matrix)
    leb =  pd.DataFrame(labels)
    # Save DataFrame to CSV
    corre.to_csv(f'/home/shir/Documents/connectomics/sch200_tian1 (copy)/HC/func/{i[-23:-16]}-cor.csv', index=False, header=False)
###    my_data = pd.concat([my_data, corre])

leb.to_csv(f'/home/shir/Documents/connectomics/sch200_tian1 (copy)/HC/leb/sch200-msai-leb.csv', index=False,header=False)
```
