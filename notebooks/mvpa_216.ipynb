function for removing outliers from dataframes' columns

```python
def replace_outliers(series, method='z', threshold=3):
    """
    This function will take a pd series and replace outliers with NaNs.
    :param series: the series input.
    :param method: method to detect outliers, 'z' for Z-score based or 'iqr' for quartiles based, default='z'.
    :param threshold: threshold for outlier detection if using the 'z' score method, default=3.
    :return: the input series with outliers replaced by NaNs.
    """
    if method == 'iqr':
        Q1 = series.quantile(0.25)
        Q3 = series.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR                
        outliers = (series < lower_bound) | (series > upper_bound) # Identify and count outliers
        num_outliers = outliers.sum()
        # Replace outliers with NaNs
     #   series.loc[outliers] = np.nan
        
    elif method == 'z':
        z_scores = ((series - series.mean()) / series.std()).abs()
        outliers = z_scores > threshold
        num_outliers = outliers.sum()
        # Replace outliers with NaNs
        series.loc[outliers] = series.mean()
        
    else:
        raise ValueError(f"Unsupported method: {method}. Use 'z' or 'iqr'.")
    return series 
```

```python
# Function to make column names unique
def make_unique(columns):
    counts = {}
    new_columns = []
    for col in columns:
        if col in counts:
            counts[col] += 1
            new_columns.append(f"{col}_{counts[col]}")
        else:
            counts[col] = 0
            new_columns.append(col)
    return new_columns

clean_data.columns = make_unique(clean_data.columns)
clean_data
```
